## Lambda Functions — 10 Examples

```
add = lambda a, b: a + b
print(add(2, 3))
```

Anonymous function that adds two numbers.
Defined in a single line using lambda.

```
square = lambda x: x*x
print(square(4))
```

Computes square of a number.
No need for def keyword.

```
even = lambda n: n % 2 == 0
print(even(6))
```

Returns True if number is even.
Useful for quick checks.

```
nums = [1,2,3]
print(list(map(lambda x: x*2, nums)))
```

Applies lambda to each list element.
Map returns transformed values.

```
nums = [1,2,3,4]
print(list(filter(lambda x: x%2==0, nums)))
```

Filters only even numbers.
Filter keeps elements meeting condition.

```
pairs = [(1,3),(2,1),(5,2)]
pairs.sort(key=lambda x: x[1])
print(pairs)
```

Sorts list using second element.
Lambda used as key function.

```
mul = lambda a,b,c: a*b*c
print(mul(2,3,4))
```

Lambda with multiple arguments.
Returns product of inputs.

```
f = lambda: "Hello"
print(f())
```

Lambda with no parameters.
Returns a constant value.

```
nums = [5,1,4]
print(sorted(nums, key=lambda x: -x))
```

Sorts list in descending order.
Negative key reverses order.

```
length = lambda s: len(s)
print(length("Python"))
```

Returns length of string.
Simple one-line function.

## Recursion — 10 Examples

```
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)
print(fact(5))
```

Computes factorial recursively.
Function calls itself with smaller input.

```
def sum_n(n):
    if n == 0:
        return 0
    return n + sum_n(n-1)
print(sum_n(5))
```

Finds sum of first n numbers.
Adds current value each call.

```
def count_down(n):
    if n == 0:
        return
    print(n)
    count_down(n-1)
count_down(3)
```

Prints numbers in reverse order.
Stops when base case reached.

```
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
print(fib(6))
```

Computes Fibonacci number.
Uses two recursive calls.

```
def power(a, b):
    if b == 0:
        return 1
    return a * power(a, b-1)
print(power(2,3))
```

Calculates exponent recursively.
Multiplies base repeatedly.

```
def reverse(s):
    if s == "":
        return ""
    return reverse(s[1:]) + s[0]
print(reverse("abc"))
```

Reverses a string using recursion.
Moves first char to end each call.

```
def gcd(a,b):
    if b == 0:
        return a
    return gcd(b, a%b)
print(gcd(12,8))
```

Finds greatest common divisor.
Uses Euclid's recursive method.

```
def digits(n):
    if n < 10:
        return 1
    return 1 + digits(n//10)
print(digits(1234))
```

Counts digits of a number.
Removes last digit each step.

```
def print_list(lst):
    if not lst:
        return
    print(lst[0])
    print_list(lst[1:])
print_list([1,2,3])
```

Prints list elements recursively.
Processes one item per call.

```
def sum_list(lst):
    if not lst:
        return 0
    return lst[0] + sum_list(lst[1:])
print(sum_list([1,2,3]))
```

Calculates sum of list elements.
Adds first element each step.
